```python
import numpy as np

def linear_func(theta, x):

    return np.sum(theta * x)
    
def linear_func_all(theta, X):

    return np.sum(theta * X, axis=1)
    
def mean_squared_error(theta, X, y):
    predict_value = linear_func_all(theta, X)

    return np.mean((predict_value - y) ** 2)
    
def grad_mean_squared_error(theta, X, y):
    predict_value = linear_func_all(theta, X)

    return 2/len(y) * np.dot(X.T, (predict_value - y))
```
